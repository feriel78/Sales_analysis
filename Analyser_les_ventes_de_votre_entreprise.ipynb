## Le mois où nous avons réalisé le meilleur chiffre d'affaires


```python
all_data
```

```python
def month(x):
    return x.split('/')[0]
```

```python
all_data['Month'] = all_data['Order Date'].apply(month)
all_data
```

```python
all_data['Month'].unique()
```

```python
all_data = all_data[all_data['Month']!='Order Date']
all_data['Month'].unique()
```

```python
all_data.dtypes
```

```python
all_data['Month']=all_data['Month'].astype(int)
all_data.dtypes
```

```python
all_data['Price Each']=all_data['Price Each'].astype(float)
all_data['Quantity Ordered']=all_data['Quantity Ordered'].astype(float)
all_data.dtypes
```

```python
all_data['Sales']=all_data['Quantity Ordered']*all_data['Price Each']
all_data
```

```python
all_data.groupby('Month')['Sales'].sum()
```

```python
months = range(1,13)
plt.bar(months,all_data.groupby('Month')['Sales'].sum())
plt.xticks(months)
plt.ylabel('Sales in USD')
plt.xlabel('Month number')
plt.show()
```

```python
products = all_data.groupby('Product')['Quantity Ordered'].sum().index
quantity = all_data.groupby('Product')['Quantity Ordered'].sum()
prices = all_data.groupby('Product')['Price Each'].mean()
```

```python
plt.figure(figsize=(40,24))
fig,ax1= plt.subplots()
ax2=ax1.twinx()
ax1.bar(products,quantity,color='g')
ax2.plot('products','prices','b-')
ax1.set_xticklabels('products',rotation='vertical',size=8)
```

```python
#2ème méthode
# concatenating
result = pd.concat([quantity, prices], axis=1)
result


# Create a figure and axis for the bar chart
fig, ax1 = plt.subplots(figsize=(8, 5))

# Bar chart on the left y-axis
ax1.bar(result.index, result['Quantity Ordered'], label='Bar Chart', color='blue', alpha=0.7)
ax1.set_xlabel('Product List')

# Rotate the x-axis labels vertically
plt.xticks(rotation='vertical')

ax1.set_ylabel('Quantity Ordered', color='blue')
ax1.tick_params('y', colors='blue')

# Create a second y-axis for the line plot on the right side
ax2 = ax1.twinx()
ax2.plot(result.index, result['Price Each'], label='Line Plot', color='red', marker='o')
ax2.set_ylabel('price for each', color='red')
ax2.tick_params('y', colors='red')

# Add title
plt.title('Bar Chart and Line Plot Example')

# Add legend
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper left')

# Show the plot
plt.show()
```
